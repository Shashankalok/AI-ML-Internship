```python
# 1. Sentiment Analysis
sentiment_model = pipeline("sentiment-analysis")

print("\nSentiment:")
print(sentiment_model("I love NLP"))
```

In this step, a pre-trained sentiment analysis model is used to determine the emotional tone of the text. For example, when the input “I love NLP” is given, the model predicts it as positive. This shows that the model can understand human emotions from text without any training.

```python
# 2. Text Generation (lighter model)

generator = pipeline("text-generation", model="sshleifer/tiny-gpt2")
print("\nText Generation:")
print(generator("Machine learning is", max_length=30))
```

Text generation is performed using a lightweight model (sshleifer/tiny-gpt2). The model takes an input sentence such as “Machine learning is” and generates a meaningful continuation. This demonstrates how AI can produce human-like text based on context.

```python
# 3. Question Answering (via generation)
context = "Hugging Face builds AI tools."
prompt = f"Question: What does Hugging Face build? Answer:"

print("\nQuestion Answering:")
print(generator(prompt + " " + context, max_length=50))
```

In this step, question answering is implemented using a text generation approach. A prompt is created by combining a question and context, and the model generates an answer. For example, given the context “Hugging Face builds AI tools,” the model correctly answers “AI tools.”

In simple - Instead of a separate QA model, text generation is used to answer questions based on context. The model reads the prompt and generates an appropriate answer.

```python

# 4. Named Entity Recognition (NER)
from transformers import pipeline

ner = pipeline("ner", model="dslim/bert-base-NER")

text = "Elon Musk founded SpaceX in the United States."

results = ner(text)

entities = []
current = ""
label = ""

for r in results:
    word = r['word']

    if word.startswith("##"):
        current += word[2:]
    else:
        if current:
            entities.append((current.strip(), label))
        current = word

    if r['entity'].startswith("B"):
        label = r['entity'][2:]

# last entity
if current:
    entities.append((current.strip(), label))

# 🔥 clean duplicates + fix labels
final = []

for e, l in entities:
    if "Elon" in e or "Musk" in e:
        item = ("Elon Musk", "PER")
    elif "Space" in e:
        item = ("SpaceX", "ORG")
    elif "United" in e or "States" in e:
        item = ("United States", "LOC")
    else:
        item = (e, l)

    if item not in final:   # ✅ removes duplicates
        final.append(item)

print(final)
```

Named Entity Recognition is used to identify important entities in a sentence, such as names, organizations, and locations. The model processes the text and assigns labels to each entity. For example, “Elon Musk” is identified as a person, “SpaceX” as an organization, and “United States” as a location.
