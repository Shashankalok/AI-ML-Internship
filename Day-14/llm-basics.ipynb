```python
from groq import Groq

client = Groq(api_key="groqapi")

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Explain AI in simple words"}
    ]
)

print(response.choices[0].message.content)
```

```python
# Temperature Experiment
response_low = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": "Write a short story"}],
    temperature=0.2
)

response_high = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": "Write a short story"}],
    temperature=1.0
)

print("Low Temperature:\n", response_low.choices[0].message.content)
print("\nHigh Temperature:\n", response_high.choices[0].message.content)
```

```python
# System vs User Prompt
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "system", "content": "You are a helpful teacher"},
        {"role": "user", "content": "Explain machine learning"}
    ]
)

print(response.choices[0].message.content)
```

```python
# Zero-shot Prompting
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Translate 'Hello' into Hindi"}
    ]
)

print(response.choices[0].message.content)
```

```python
# Few-shot Prompting
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": """
English: Hello → Hindi: नमस्ते
English: Thank you → Hindi: धन्यवाद
English: Good morning →
"""}
    ]
)

print(response.choices[0].message.content)
```

```python
# Chain-of-Thought Prompting
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Solve step-by-step: If a train travels 60 km in 1 hour, how far in 3 hours?"}
    ]
)

print(response.choices[0].message.content)
```

```python
# Summarization Prompt
text = """Artificial Intelligence is transforming industries by enabling machines to learn from data and make decisions."""

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": f"Summarize this in 2 bullet points:\n{text}"}
    ]
)

print(response.choices[0].message.content)
```

```python
# Extraction Prompt
text = "John met Mary on 12th June 2023 in New York."

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": f"Extract names and dates:\n{text}"}
    ]
)

print(response.choices[0].message.content)
```

```python
# Generation Prompt (Email)
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Write a professional email requesting leave for 2 days"}
    ]
)

print(response.choices[0].message.content)
```

```python
# Multiple Prompt Variations
prompts = [
    "Explain machine learning simply",
    "Explain machine learning like I am 10",
    "Explain ML in bullet points",
    "Explain ML with examples",
    "Explain ML briefly",
]

for p in prompts:
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": p}]
    )
    print(f"\nPrompt: {p}")
    print(response.choices[0].message.content)
```

```python
# Prompt Comparison (Structured vs Basic)
basic = "Explain AI"

structured = "Explain Artificial Intelligence in simple words with examples and in 3 bullet points"

res1 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": basic}]
)

res2 = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": structured}]
)

print("Basic:\n", res1.choices[0].message.content)
print("\nStructured:\n", res2.choices[0].message.content)
```

```python
# Cost Calculation
input_tokens = 500
output_tokens = 500
price_per_1000_tokens = 0.002

total_cost = ((input_tokens + output_tokens) / 1000) * price_per_1000_tokens

print("Estimated Cost: $", total_cost)
```

```python
# Prompt Template Library (Save File)

templates = """
Summarization:
Summarize the following text in {n} bullet points:

Extraction:
Extract {specific information} from the text:

Generation:
Act as a {role} and generate {output}:

Reasoning:
Solve step-by-step:

Email:
Write a professional email for {purpose}:
"""

with open("prompt_templates.txt", "w") as f:
    f.write(templates)

print("Prompt template file created")
```

```python
# stand comedy prompt
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {
            "role": "user",
            "content": """
Example:
"Yaar alarm ka relationship mujhse zyada toxic hai… main band karta hoon, wo phir bajta hai!"

Now write a similar funny stand-up comedy script about sleep in Hinglish.
"""
        }
    ],
    temperature=1.0
)

print(response.choices[0].message.content)
```

```python
# song prompt
from groq import Groq

client = Groq(api_key="groqapi")

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {
            "role": "system",
            "content": "You are a professional Indian hip-hop lyricist who writes in Hinglish with strong rhymes, rhythm, and street-style vibe."
        },
        {
            "role": "user",
            "content": """
Write a full Indian hip-hop song in Hinglish.

Requirements:
- Style: Desi hip-hop (street vibe, energetic, rhythmic)
- Language: Mix of Hindi + English (Hinglish)
- Structure:
    1. Intro
    2. Verse 1
    3. Hook/Chorus (catchy and repeatable)
    4. Verse 2
    5. Bridge
    6. Final Hook
- Theme: Struggles, hustle, dreams, and lifestyle
- Include rhyming lines and strong punchlines
- Make it sound like a real performance song

Keep it engaging, rhythmic, and expressive.
"""
        }
    ],
    temperature=0.9
)

print(response.choices[0].message.content)
```
