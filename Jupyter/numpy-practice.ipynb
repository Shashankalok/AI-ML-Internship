Numpy

```python
import numpy as np
```

```python
##Creating a 1D Array (Vector)
#A NumPy array is like a vector in mathematics or R.

a = np.array([1, 2, 3, 4])
print(a)
```

```python
##Creating a 2D Array (Matrix)
#A 2D array represents a matrix with rows and columns.

b = np.array([[1, 2], [3, 4]])
print(b)
```

```python
##Mathematical Operations
#Perform element-wise addition and multiplication.
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(x + y)
print(x * y)

np.dot(x, y)
int(np.dot(x, y))
```

```python
##Broadcasting
#Add a scalar to all elements of an array.
x = np.array([1, 2, 3])

print(x + 5)
```

```python
##Statistical Functions
#Calculate mean, variance, and standard deviation.

data = np.array([2, 4, 6, 8])

print(np.mean(data))
print(np.var(data))
print(np.std(data))
```

```python
##Indexing
#Access specific elements (Python uses 0-based indexing).

x = np.array([10, 20, 30, 40])

print(x[0])
print(x[2])

#Negative Indexing
x = np.array([10, 20, 30, 40])

print(x[-1])
print(x[-2])
```

```python
##Slicing
#Extract a subset of elements from an array.
x = np.array([10, 20, 30, 40, 50])

print(x[1:4])
```

```python
##Matrix Indexing (2D Indexing)
#Access elements from a 2D array using row and column indices.

A = np.array([[1, 2], [3, 4]])

print(A[0, 1])
```

```python
##3D Indexing (Tensor)

arr = np.arange(27)
arr_3d = arr.reshape(3,3,3)

print(arr_3d[2,0,2])
```

```python
##Reshaping Arrays
#Change the shape of an array (vector to matrix).

x = np.array([1, 2, 3, 4, 5, 6])

y = x.reshape(2, 3)
print(y)
```

```python
##Shape and Dimension
#Get size, shape, and dimensions of an array.

x = np.array([[1, 2, 3], [4, 5, 6]])

print(x.shape)
print(x.ndim)
print(x.size)
```

```python
##Random Number Generation
#Generate random numbers for simulations and statistical analysis.

print(np.random.rand(5))        # Uniform distribution
print(np.random.randn(5))       # Normal distribution
print(np.random.randint(1, 10, 5))
```

```python
##Random Number generation using inverse theorem

import numpy as np

U = np.random.rand(5)
X = -np.log(1 - U)

print(X)
```

```python
##PRNG

np.random.seed(42)   # set seed
print(np.random.rand(3))
```

```python
##Creating Sequences
#Generate sequences similar to R's seq() function.
print(np.arange(1, 10, 2))
print(np.linspace(0, 1, 5))
```
