```python
def calculator(operation, a, b):
    try:
        if operation == "add":
            return a + b
        elif operation == "subtract":
            return a - b
        elif operation == "multiply":
            return a * b
        elif operation == "divide":
            if b == 0:
                return "Error: Division by zero"
            return a / b
        else:
            return "Invalid operation"
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
print(calculator("multiply", 5, 10))
```

```python
calculator("add", 2, 3)
```

```python
from langchain.tools import tool

@tool
def calculator_tool(input: str) -> str:
    """
    Performs basic arithmetic operations.
    Input should be in format: operation,a,b
    Example: add,2,3
    """
    try:
        operation, a, b = input.split(",")
        a, b = float(a), float(b)

        if operation == "add":
            return str(a + b)
        elif operation == "subtract":
            return str(a - b)
        elif operation == "multiply":
            return str(a * b)
        elif operation == "divide":
            if b == 0:
                return "Error: Division by zero"
            return str(a / b)
        else:
            return "Invalid operation"
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
print(calculator_tool.run("multiply,5,10"))
```

```python
print(calculator_tool.run("divide,10,0"))
```

```python
print(calculator_tool.run("add,10,5"))
```

```python
if llm_response.get("function_call"):
    func_name = llm_response["function_call"]["name"]
    args = llm_response["function_call"]["arguments"]

    if func_name == "calculator":
        result = calculator(**args)
        print("Result:", result)
```
