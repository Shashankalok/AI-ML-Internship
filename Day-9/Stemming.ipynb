```python
import nltk
nltk.download('punkt') 
```

```python
words = ["eating","eats","eaten","writing","writes",
         "programming","programs","history","finally","finalized"]
```

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()

for word in words:
    print(word + " ----> " + stemming.stem(word))
```

```python
print(stemming.stem('congratulations'))
print(stemming.stem('sitting'))
```

```python
from nltk.stem import RegexpStemmer

reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)

print(reg_stemmer.stem('eating'))
print(reg_stemmer.stem('ingeating'))
```

```python
from nltk.stem import SnowballStemmer

snowballstemmer = SnowballStemmer('english')

for word in words:
    print(word + " ----> " + snowballstemmer.stem(word))
```

```python
print("Porter:", stemming.stem("fairly"), stemming.stem("sportingly"))
print("Snowball:", snowballstemmer.stem("fairly"), snowballstemmer.stem("sportingly"))

print("Snowball goes:", snowballstemmer.stem("goes"))
print("Porter goes:", stemming.stem("goes"))
```
