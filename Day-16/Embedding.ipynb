```python
from sentence_transformers import SentenceTransformer

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# sentences
sentences = [
    "I love cricket",
    "I enjoy sports",
    "I hate traffic",
    "Bangalore has heavy traffic"
]

# Create embeddings
embeddings = model.encode(sentences)

# Print results
for i, emb in enumerate(embeddings):
    print(f"Sentence: {sentences[i]}")
    print(f"Embedding (first 5 values): {emb[:5]}")
    print("------")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compare similar sentences
sim1 = cosine_similarity([embeddings[0]], [embeddings[1]])
print("Cricket vs Sports:", sim1)

# Compare traffic sentences
sim2 = cosine_similarity([embeddings[2]], [embeddings[3]])
print("Traffic vs Bangalore Traffic:", sim2)
```

```python
import chromadb
from sentence_transformers import SentenceTransformer

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Sample data
documents = [
    "I love cricket",
    "I enjoy sports",
    "I hate traffic",
    "Bangalore has heavy traffic"
]

# Convert to embeddings
embeddings = model.encode(documents)

# Initialize ChromaDB
client = chromadb.Client()

# Create collection
collection = client.create_collection(name="my_collection")

# Add data to DB
collection.add(
    documents=documents,
    embeddings=embeddings.tolist(),
    ids=["1", "2", "3", "4"]
)

print("Data stored successfully!")
```

```python
query = "sports"

results = collection.query(
    query_texts=[query],
    n_results=2
)

print(results)
```

```python
queries = ["traffic", "games", "city problems"]

for q in queries:
    results = collection.query(query_texts=[q], n_results=2)
    print(f"\nQuery: {q}")
    print(results['documents'])
```
